```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
# from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

import pandas as pd

# data = pd.read_csv('answers.csv')
data = pd.read_csv('questions.csv')

# print(data['answers_body'])

# newsgroups = fetch_20newsgroups(remove=('headers', 'footers', 'quotes'))
# docs_raw = newsgroups.data

# print(docs_raw[0])

docs_raw = data['questions_body'].values.astype('U')




tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5,
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(docs_raw)
print(dtm_tf.shape)


tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)


# for TF DTM
#lda = LatentDirichletAllocation(n_topics=20, random_state=0)
#lda.fit(dtm_tf)
# for TFIDF DTM
lda = LatentDirichletAllocation(n_topics=20, random_state=0)
lda.fit(dtm_tfidf)

pyLDAvis.sklearn.prepare(lda, dtm_tf, tf_vectorizer)

```
