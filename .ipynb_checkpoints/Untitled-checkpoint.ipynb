```python
answer_scores.set_index('id', inplace=True, drop=True)
answers.set_index('answers_id', inplace=True, drop=True)
answers_w_scores = answers.join(other=answer_scores, on='answers_id', how='inner')
columns = ['answers_author_id', 'answers_question_id', 'answers_date_added', 'answers_body', 'score']
answers_w_scores = answers_w_scores[columns]
answers_w_scores.head()
answers_w_scores.to_csv('clean_data/answers_w_scores.csv')
```
